```python
import pandas as pd
import requests

# URL de la API DIVIPOLA de datos.gov.co
url = "https://www.datos.gov.co/resource/gdxc-w37w.json?$limit=1000"

# Hacer la petición
response = requests.get(url)
response.raise_for_status()  # Lanza error si no se pudo conectar

# Convertir JSON a DataFrame
data = response.json()
df = pd.DataFrame(data)

# Mostrar las primeras filas y nombres de columnas
print("Columnas disponibles:")
print(df.columns.tolist())
df.head()
```
